```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, cos, sin, asin, sqrt

def haversine(coord1, coord2):
    # Convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [coord1[1], coord1[0], coord2[1], coord2[0]])
    
    # Debugging: Print values to check
    print(f"Coordinates 1: {coord1}, Coordinates 2: {coord2}")
    
    # Haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371  # Radius of Earth in kilometers
    distance = c * r
    
    # Debugging: Print intermediate result
    print(f"Distance (in km): {distance}")
    return distance

# Example usage:
launch_site = [28.562302, -80.577356]  # CCAFS LC-40
proximity_site = [28.563197, -80.576820]  # Nearby point

distance_km = haversine(launch_site, proximity_site)
print(f"Distance: {distance_km:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import radians, sin, cos, sqrt, atan2
import folium
from folium import Marker
from folium.features import DivIcon
from folium.plugins import MousePosition

# Function to calculate distance using the Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of the Earth in kilometers
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordinates of the Launch Site
launch_site_lat = 28.562341
launch_site_lon = -80.57678

# Coordinates for nearby locations (updated for better accuracy)
city_lat = 28.6122  # Titusville, FL
city_lon = -80.8081

railway_lat = 28.5740  # Adjusted coordinates for NASA Railway (close to the launch site)
railway_lon = -80.6483  # Adjusted coordinates

highway_lat = 28.5735  # SR 405 (Highway)
highway_lon = -80.6480

# Calculate distances from the launch site to the closest city, railway, and highway
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Print the distances to the output
print(f"Distance from Launch Site to City (Titusville): {distance_city:.2f} km")
print(f"Distance from Launch Site to Railway (NASA Railroad): {distance_railway:.2f} km")
print(f"Distance from Launch Site to Highway (SR 405): {distance_highway:.2f} km")

# Initialize the map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)  # Zoom in to better view the markers

# Add a marker for the launch site
Marker([launch_site_lat, launch_site_lon], popup='Launch Site').add_to(site_map)

# Add a marker for the city (Titusville)
Marker([city_lat, city_lon], popup=f'City (Titusville): {distance_city:.2f} km', icon=folium.Icon(color='purple')).add_to(site_map)

# Add a marker for the railway station (NASA Railroad) with a different icon
Marker([railway_lat, railway_lon], popup=f'NASA Railroad: {distance_railway:.2f} km', icon=folium.Icon(color='red', icon='cloud')).add_to(site_map)

# Add a marker for the highway (SR 405) with a different icon
Marker([highway_lat, highway_lon], popup=f'SR 405 (Highway): {distance_highway:.2f} km', icon=folium.Icon(color='green')).add_to(site_map)

# Draw a line between the launch site and the city
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]],
    color='purple',
    weight=2
).add_to(site_map)

# Draw a line between the launch site and the railway station (NASA Railroad)
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]],
    color='red',
    weight=2
).add_to(site_map)

# Draw a line between the launch site and the highway (SR 405)
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]],
    color='green',
    weight=2
).add_to(site_map)

# Add distance labels at the midpoints of the lines
folium.Marker(
    location=[(launch_site_lat + city_lat) / 2, (launch_site_lon + city_lon) / 2],
    icon=DivIcon(
        html=f'<div style="font-size: 12px; color: purple;"><b>{distance_city:.2f} km</b></div>'
    )
).add_to(site_map)

folium.Marker(
    location=[(launch_site_lat + railway_lat) / 2, (launch_site_lon + railway_lon) / 2],
    icon=DivIcon(
        html=f'<div style="font-size: 12px; color: red;"><b>{distance_railway:.2f} km</b></div>'
    )
).add_to(site_map)

folium.Marker(
    location=[(launch_site_lat + highway_lat) / 2, (launch_site_lon + highway_lon) / 2],
    icon=DivIcon(
        html=f'<div style="font-size: 12px; color: green;"><b>{distance_highway:.2f} km</b></div>'
    )
).add_to(site_map)

# Enable MousePosition plugin to display coordinates on mouse hover
MousePosition().add_to(site_map)

# Display the final map
site_map
```
